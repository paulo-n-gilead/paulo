# Terminal Integration Test Notebook
Checks for JupyterLab terminal support (`terminado` + `jupyter_server_terminals`) and a subprocess shell demo.

```python
import importlib
mods = ['terminado', 'jupyter_server_terminals']
for m in mods:
    mod = importlib.import_module(m)
    print(f'{m}: OK ({getattr(mod, "__version__", "version-unknown")})')
```

```python
import subprocess
result = subprocess.run(['bash', '-lc', 'echo Terminal subprocess OK && pwd && whoami'], capture_output=True, text=True, check=True)
print(result.stdout)
```

```python
from jupyter_server.serverapp import ServerApp
traits = list(ServerApp.class_trait_names())
terminal_related = [t for t in traits if 'terminal' in t.lower()]
print('ServerApp terminal-related traits:')
for t in sorted(terminal_related):
    print(' -', t)
```
