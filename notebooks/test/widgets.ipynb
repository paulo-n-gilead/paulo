# Widgets Test Notebook
Checks `ipywidgets` and `nglview` availability, then runs interactive demos.

```python
import ipywidgets as widgets
import nglview as nv
print('ipywidgets version:', widgets.__version__)
print('nglview version:', nv.__version__)
```

```python
slider = widgets.IntSlider(value=5, min=0, max=10, step=1, description='Value')
display(slider)
slider
```

```python
text = widgets.Text(value='hello', description='Input')
output = widgets.Output()
def on_change(change):
    with output:
        output.clear_output()
        print('Current text:', change['new'])
text.observe(on_change, names='value')
display(widgets.VBox([text, output]))
```
